```python
import pandas as pd

edges_data = pd.read_excel(linkGoogle, sheet_name='edges')
```

```python
# read in attributes
attr_data = pd.read_excel(linkGoogle, sheet_name='attributes')

import numpy as np

# attr_data['followers'] = np.log(attr_data['followers'])

attr_data
```

```python
edges_data
```

```python
import networkx as nx

graph = nx.from_pandas_edgelist(edges_data, edge_attr = True,
                                         create_using = nx.DiGraph) # here!!

# see edges
graph.edges.data()
```

```python
# add attributes

attr_male = {n:s for n,s in zip(attr_data.name,attr_data.male)}
attr_followers = {n:s for n,s in zip(attr_data.name,attr_data.followers)}

nx.set_node_attributes(graph, attr_male , "male")
nx.set_node_attributes(graph, attr_followers , "followers")
```

```python
graph.nodes.data()
```

```python


colorNodes = {node:'blue' if male==1 else 'pink' for node, male in nx.get_node_attributes(graph, "male").items()}

# node_colors = nx.get_node_attributes(graph, "male").values()
nx.set_node_attributes(graph, colorNodes, "color")

node_colors = nx.get_node_attributes(graph, "color").values()

node_sizes = nx.get_node_attributes(graph, "followers").values()

# Scale node_sizes to a reasonable range
max_followers = max(node_sizes)
size = [size * 1000 / max_followers for size in node_sizes]
```

```python
nx.draw(graph,
        with_labels=True,
        node_color = node_colors,
        node_size = list(size), arrowsize = 20)
```

```python
from google.colab import files
folder = "" # this folder needs to be created previously on your machine

nx.write_graphml(graph, folder+"seattle.graphml",named_key_ids=True)

files.download(folder+"seattle.graphml")
```
